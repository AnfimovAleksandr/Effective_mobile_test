# Задача 1: Изоморфизм

Решение 1: Сложность по времени O(n), сложность по памяти O(k) ~ O(1) (из-за ограниченности алфавита, k - количество уникальных символов)

```python
def is_isomorphic(s, t):
    n = len(s)
    m = len(t)
    if m != n:
        return False
    swaps_s = {}
    swaps_t = {}
    for i in range(n):
        if s[i] in swaps_s:
            if swaps_s[s[i]] != t[i]:
                return False
        else: 
            swaps_s[s[i]] = t[i]
        if t[i] in swaps_t:
            if swaps_t[t[i]] != s[i]:
                return False
        else: 
            swaps_t[t[i]] = s[i]
    return True

# Пример:
s = 'paper' 
t = 'title' 
print(is_isomorphic(s, t))
```

Решение 2: Красиво в одну строчку, но медленнее в среднем, так что чисто для красоты

С почти той же ассимптотикой (3*O(n) ~ O(n) по времени и O(r) + O(k) + O(k) ~ O(1) по памяти (из-за ограниченности алфавита, r - количество уникальных пар символов (s, t), k - количество уникальных символов))

```python
def is_isomorphic(s, t):
    return len(set(s)) == len(set(t)) == len(set(zip(s,t)))
# Пример:
s = 'paper' 
t = 'title' 
print(is_isomorphic(s, t))
```

# Задача 2: Натуральная последовательность

Сложность: O(log(n)) по времени и O(1) по памяти

```python
def missing_number(nums):
    n = len(nums)
    left, right = 0, n - 1
    
    if n == 0 or nums[0] == 2:
        return 1
    if nums[-1] == n:
        return n + 1
    
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] > mid + 1:
            ans = mid + 1 
            right = mid - 1
        else:
            left = mid + 1
    
    return ans

# Пример:
nums = [1, 2, 3, 4, 5, 6, 8, 9, 10, 11]
print(missing_number(nums))
```

# Задание 3: Факторизация

Сложность: O($\sqrt{n}$) по времени и O(log(n)) по памяти

```python
def prime_factors(n):
    if n <= 1:
        return []
    
    factors = []

    while n % 2 == 0:
        factors.append(2)
        n //= 2
    
    divisor = 3
    while divisor * divisor <= n:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 2
    
    if n > 1:
        factors.append(n)
    
    return factors

# Пример:
n = 56
print(prime_factors(n))
```
