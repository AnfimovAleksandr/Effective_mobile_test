# Задача 1: Абитуриенты

```python
SELECT
    id,
    scores,
    RANK() OVER (ORDER BY scores DESC) AS rating
FROM examination
WHERE scores IS NOT NULL
ORDER BY rating
```

# Задача 3: Покупки

```python
SELECT a.client_id
FROM account a
JOIN transaction t ON a.id = t.account_id
WHERE 
    t.transaction_date >= CURRENT_DATE - INTERVAL '1 month'
    AND t.type = 'BUY' -- тип взял с потолка
GROUP BY a.client_id
HAVING SUM(t.amount) < 5000;
```
